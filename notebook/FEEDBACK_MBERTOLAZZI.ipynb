````python
# Replicate this figure
# Temperature values are originally in Kelvin units -> Convert to degree celsius by subtracting 273.15
fig, (g1,g2) = plt.subplots(1, 2, figsize=(12,5),gridspec_kw={'width_ratios': [5, 1.5]})
ctemp = g1.contourf(lon,lat,temp - 273.15,cmap='magma',levels=np.linspace(-30,40,15),extend='both')
g1.contour(lon,lat,temp - 273.15,colors='w',levels=[-10],extend='both')
g1.set_xlabel('Longitude')
g1.set_ylabel('Latitude')
g1.set_title('Current Global Temperature')
fig.colorbar(ctemp,ax=g1,label='$^o$C')

g2.plot(np.mean(temp-273.15,axis=1),lat,lw=2,c='k')
#n the context of this data, where temp likely has dimensions corresponding to lat
#and long, axis=1 would correspond to the long dimension.
#Therefore, np.mean(temp-273.15,axis=1) calculates the average temperature across all long
#for each lat, giving you the zonal mean temperature.
g2.set_xlabel(r'$^{o}$C')
g2.set_ylabel(r'Latitude')
g2.set_title('Zonal Mean Temperature')
g2.grid()
plt.show()
````

````python
# This is the function to the set the ticks in the colorbar to be in scientific notations.
# You can directly use this function and don't need to change anything in this cell.
import matplotlib.ticker as ticker
def fmt(x, pos):
    a, b = '{:.2e}'.format(x).split('e')
    b = int(b)
    return r'$10^{{{}}}$'.format(a)
````

````python
# Replicate the figure here
fig,eth1 = plt.subplots(1, 1, figsize=(22,14))
ctemp=eth1.scatter(longitude,latitude,s=magnitude**2,c=np.log10(depth),marker='o',cmap='viridis',vmin=0,vmax=2.5)
# magnitude**2 per visualizzare in modo migliore la differenza di taglia legata alla magnitudo
eth1.set_xlabel('Longitude($^{o}$)')
eth1.set_ylabel('Latitude($^{o}$)')
plt.grid()
fig.colorbar(ctemp,ax=eth1,label='Depth [m]',format=ticker.FuncFormatter(fmt))
eth1.set_title('Location of significant earthquakes with size indicating relative magnitude')
plt.show()
````

````python
#SOLUTION

import matplotlib.pyplot as plt
#import cartopy.feature as cfeature
import numpy as np

fig,ax = plt.subplots(1,2,figsize=(18, 12),subplot_kw={'projection': ccrs.Orthographic(central_latitude=-90, central_longitude=0)})
# The rest doesn't change:
ax[0].set_extent([-180, 180, -90, -55], ccrs.PlateCarree())

######################################################################################################################################################
# Add Land and gridlines
# Hint: Check edgecolor to show coast, and facecolor to color the continent
# Hint: https://matplotlib.org/stable/gallery/color/named_colors.html for the names of the colors available
######################################################################################################################################################
ax[0].add_feature(cartopy.feature.LAND, edgecolor='black',facecolor='silver')
ax[0].gridlines(lw=1,color='w')
######################################################################################################################################################
# Add Contours
# Hint: X (Y) data in the pcolormesh should be the longitude and latitude of the sea ice data.
# Hint: Check the shape of the var_toplot1, var_toplot2 before plotting
# Hint: longitude and latitude can be accessed with seaice1.variables['longitude'][:], seaice1.variables['latitude'][:]
# Hint: Check this website: https://matplotlib.org/stable/tutorials/colors/colormaps.html for available colormaps
######################################################################################################################################################
ICE1 = ax[0].pcolormesh(seaice1.variables['longitude'][:], seaice1.variables['latitude'][:], np.ma.masked_greater(var_toplot1[0, :, :],1), transform=ccrs.PlateCarree(),vmin=0,vmax=1,cmap='Blues')
plt.colorbar(ICE1,ax=ax[0],fraction=0.046, pad=0.04,label='Sea Ice Concentration')
ax[0].set_title('2017-08-01')

ax[1].set_extent([-180, 180, -90, -55], ccrs.PlateCarree())
ax[1].add_feature(cartopy.feature.LAND, edgecolor='black',facecolor='silver')
ax[1].gridlines(lw=1,color='w')

ICE2 = ax[1].pcolormesh(seaice2.variables['longitude'][:], seaice2.variables['latitude'][:], np.ma.masked_greater(var_toplot2[0,:,:],1), transform=ccrs.PlateCarree(),vmin=0,vmax=1,cmap='Blues')
plt.colorbar(ICE2,ax=ax[1],label='Sea Ice Concentration',fraction=0.046, pad=0.04)
ax[1].set_title('2017-12-31')
plt.show()

````
